# Visualization of the behavior of the hyperbolic tangent function

The objective of this code is to graph the hyperbolic tangent function for different values of the geometric parameter $\alpha$, and the transition slope $k_{p}$.

\begin{equation} 
S(\alpha,k_{p}) = \frac{1}{2}-\frac{1}{2}tanh\left (k_{p}\left (\alpha-\frac{1}{2} \right )\right ) \quad \quad \alpha \in  [ 0, 1 ]  \quad \quad k_{p} \in [ 0, +\infty [ 
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
%matplotlib inline
plt.rcParams["figure.figsize"] = (16, 12)
plt.rcParams['xtick.labelsize'] = 24
plt.rcParams['ytick.labelsize'] = 24
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
```

```python
alfa = np.arange(0,1,0.005)
S0 = 0.5-0.5*np.tanh(0*(alfa-0.5))
S1 = 0.5-0.5*np.tanh(1*(alfa-0.5))
S3 = 0.5-0.5*np.tanh(3*(alfa-0.5))
S6 = 0.5-0.5*np.tanh(6*(alfa-0.5))
S14 = 0.5-0.5*np.tanh(14*(alfa-0.5))
S100 = 0.5-0.5*np.tanh(600*(alfa-0.5))
Sinf = 0.5-0.5*np.tanh((10**10)*(alfa-0.5))

plt.plot(alfa, S0, c='k',linestyle=":",label='$k_{p}$=0')
plt.plot(alfa, S1, c='tab:gray',linestyle=":",label='$k_{p}$=1')
plt.plot(alfa, S3, c='k',linestyle="--",label='$k_{p}$=3')
plt.plot(alfa, S6, c='tab:gray',linestyle="--",label='$k_{p}$=6')
plt.plot(alfa, S14, c='k',linestyle="-.", label='$k_{p}$=14')
plt.plot(alfa, S100, c='tab:gray',linestyle="-.", label='$k_{p}$=100')
plt.plot(alfa, Sinf, c='k',label='$k_{p}$=+Inf')
plt.legend(loc='best',prop={'size':20})
plt.xlabel('α',fontsize = 26)
plt.ylabel('S(α)',fontsize = 26)
plt.xlim([-0.00, 1.01])
plt.ylim([-0.01, 1.01])
plt.title('Hyperbolic Tangent Function',fontsize = 28)
#plt.savefig('GraficosT/FuncionTanH')
plt.show()
```
